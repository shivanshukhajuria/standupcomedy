Storing the next occuring word for each word to generate a markov chain

```python
from collections import defaultdict

def markov_chain(text):
    words = text.split(' ')  
    m_dict = defaultdict(list)
    for current_word, next_word in zip(words[0:-1], words[1:]):
        m_dict[current_word].append(next_word)
    m_dict = dict(m_dict)
    return m_dict
```

If we end the chain with a punctuation, the sentence is likely to make more sense

```python
def generate_sentence(chain, count=10):
    punctuations = ['.','!',',','?','"']
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()

    for i in range(count-1):
        word2 = random.choice(chain[word1])
        word1 = word2
        if(word2 != "♪\n♪"):
            sentence += ' ' + word2

    sentence += random.choice(punctuations)
    return(sentence)
```
